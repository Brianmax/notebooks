```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv", sep=",")
```

```python
df.info()
```

```python
# exploracion de datos
columns_drop = ["Fare", "Name"]
df.drop(columns_drop, axis="columns",inplace=True)
```

```python
df
```

```python
columns = ["Survived", "Pclass", "Sex","Age","Saboard","Paboard"]
df.columns = columns
```

```python
df
```

```python
print(df.shape)
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df,columns=["Sex"], drop_first=True)
```

```python
df.head(6)
```

```python
df.rename(columns={"Sex_male": "Sex"}, inplace=True)
```

```python
df
```

```python
x = df.drop(["Survived"], axis=1)
y = df.Survived
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=00000)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree =  DecisionTreeClassifier(max_depth=2, random_state=00000)
```

```python
tree.fit(X_train, y_train)
```

```python
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Accuracy train:",train_accuracy)
print("Accuracy test:",test_accuracy)
```
