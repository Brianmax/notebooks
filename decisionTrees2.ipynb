```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car.csv")
df
```

```python
columns = ["price","maintance","doors","persons","lug","safety","decision"]
```

```python
df.columns = columns
```

```python
x = df.drop(["decision"], axis=1)
y = df["decision"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
import category_encoders as ce
encoder = ce.OrdinalEncoder(cols=["price","maintance","doors","persons","lug","safety"])
```

```python
x_train = encoder.fit_transform(x_train)
```

```python
x_test = encoder.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=2, random_state=0)
```

```python
tree.fit(x_train, y_train)
```

```python
y_train_pred = tree.predict(x_train)
y_test_pred = tree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
train_accuracy_tree = accuracy_score(y_train, y_train_pred)
test_accuracy_tree = accuracy_score(y_test, y_test_pred)

print("Train accuracy:",train_accuracy_tree)
print("Test accuracy:",test_accuracy_tree)
```
